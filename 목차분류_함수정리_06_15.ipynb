```python
import pandas as pd
import numpy as np
import re
```

```python
# 분류 항목 베이직 버전

## 목차 수집 정보 전처리
def clear_index(a, i):
    c = a[i].translate({ord('['): None, ord(']'): None, ord("\'"):None, ord('·'):None})
    c = c.replace('목차 더보기','')  # 불필요한 요소 삭제
    c = c.split(',')
    c = [j.strip() for j in c]
    if '' in c: # 빈칸 삭제
        c.remove('') 
    return c

## 공백 기준으로 잘랐을때, 분류가 가능한 경우
def pandan_1(c):
    pd1 = []
    for ic in c:
        ic = ic.split(' ')[0]
        if ic[-1] not in pd1:
            pd1.append(ic[-1])
    if len(pd1) == 1:
        for ic in c:
            big.append(ic)
            mid.append(np.NaN)
            small.append(np.NaN)
    elif len(pd1) == 2:
        for ic in c:
            icc = ic.split(' ')[0]
            if icc[-1] == pd1[0]:
                big.append(ic)
                mid.append(np.NaN)
                small.append(np.NaN)
            elif icc[-1] == pd1[1]:
                mid.append(ic)
                big.append(np.NaN)
                small.append(np.NaN)
    elif len(pd1) >= 3:
        for ic in c:
            icc = ic.split(' ')[0]
            if icc[-1] == pd1[0]:
                big.append(ic)
                small.append(np.NaN)
                mid.append(np.NaN)
            elif icc[-1] == pd1[1]:
                mid.append(ic)
                big.append(np.NaN)
                small.append(np.NaN)
            else:
                big.append(np.NaN)
                mid.append(np.NaN)
                small.append(ic)
```

```python
# 적용 예시

## 데이터 불러오기
cate = pd.read_excel("C:\\police_data\\total_idx4.xlsx")
a = cate['index']  # 목차 컬럼 접근

## 대, 중, 소 분류
big, mid, small = [],[],[]

## 수집 전처리 및 분류
c = clear_index(a,1)
pandan_1(c)

## 데이터 프레임화
df2 = pd.DataFrame({'big':big, 'mid':mid, 'small':small})
```
